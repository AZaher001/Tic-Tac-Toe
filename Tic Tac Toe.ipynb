```python
board = {'1': ' ' , '2': ' ' , '3': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '7': ' ' , '8': ' ' , '9': ' ' }
board_again = []
for x in board:
    board_again.append(x)
```

```python
def drawboard(theboard):
    print(board['1'] + '    |' + board['2'] + '    |' + board['3'])
    print('_____|_____|_____')
    print(board['4'] + '    |' + board['5'] + '    |' + board['6'])
    print('_____|_____|_____')    
    print(board['7'] + '    |' + board['8'] + '    |' + board['9']) 
    print('     |     |     ')    
```

```python
def game():
    player1 = input('player1 , Enter X or O : ' )
    player2 = input('player2 , Enter X or O : ' )
    turn = player1
    count = 0
    for i in range(10):
        drawboard(board)
        print("It's your turn," + turn)
        move = input("Move to which place? ")
        if board[move] == ' ':
            board[move] = turn
            count += 1
        else:
            print("That place is already full . Move to other place?")
            continue

        if turn == player1:
            turn =  player2
        else:
            turn = player1
            
        if count >= 5 : 
            if board['1'] == board['2'] == board['3'] != ' ':
                drawboard(board)
                print('The End')
                print(board['1'] + 'won.')
                break
            elif board['4'] == board['5'] == board['6'] != ' ':
                drawboard(board)
                print('The End')
                print(board['4'] + 'won.')
                break
            elif board['7'] == board['8'] == board['9'] != ' ':
                drawboard(board)
                print('The End')
                print(board['7'] + 'won.')
                break
            elif board['1'] == board['4'] == board['7'] != ' ':
                drawboard(board)
                print('The End')
                print(board['1'] + 'won.')
                break
            elif board['2'] == board['5'] == board['8'] != ' ':
                drawboard(board)
                print('The End')
                print(board['2']+ 'won.')
                break
            elif board['3'] == board['6'] == board['9'] != ' ':
                drawboard(board)
                print('The End')
                print(board['3']+ 'won.')
                break
            elif board['1'] == board['5'] == board['9'] != ' ':
                drawboard(board)
                print('The End')
                print(board['1'] + 'won.')
                break
            elif board['3'] == board['5'] == board['7'] != ' ':
                drawboard(board)
                print('The End')
                print(board['3'] + 'won.')
                break
        if count == 9 :
            print('game Over . /nno winner.')
    
    while True :
        restart = input('you want to play again !! (y or n)??')
        if restart =='y' or restart =='Y' or restart =='yes':
            for again in board_again :
                board[again] = ' '
        else:
            break
            

        game()

game()
```
